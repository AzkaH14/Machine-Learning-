```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pit
import seaborn as sns
```

```python
data_frame = pd.read_csv('/content/drive/MyDrive/Asiignemnt1-ML/Automobile_data (1).csv')
```

```python
data_frame.info()
```

```python
#Show summary statistics
data_frame.describe()
```

```python
#Calculate mean
data_frame['width'].mean()
```

```python
#Calculate median
data_frame['width'].median()
```

```python
# Checking for duplicates
data_frame.duplicated()
```

```python
#Drop duplicates
data_frame.drop_duplicates()
```

```python
#Replace '?' with NaN
data_frame.replace('?', np.nan, inplace=True)
```

```python
#Checking missing values
new_data=data_frame.isnull().sum()
```

```python
#Converting 'normalized-losses' to numeric
data_frame['normalized-losses'] = pd.to_numeric(data_frame['normalized-losses'])

#Replace the missing values in 'normalized-losses' with mean of the column
mean_normalized_losses = data_frame['normalized-losses'].mean()
data_frame['normalized-losses'].fillna(mean_normalized_losses, inplace=True)

print("Imputed missing values in 'normalized-losses' with the mean.")
```

```python
new_data.info()
```

```python
#Drop rows where 'price' is NaN
data_frame = data_frame.dropna(subset=['price'])

print("Rows having missing values dropped in the price column.")
```
